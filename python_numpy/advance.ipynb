```python
import numpy as np
import matplotlib as plt
```

```python
sales_data=np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 2025000, 230000]
])


print("======= Zomato sales analysis =======")
print("\n Sales data shape: ", sales_data.shape )
print("sample data for 1st 3 restau: \n", sales_data[0:3])
```

```python
# total sales per year

print(np.sum(sales_data, axis=0))
yeary_total=np.sum(sales_data[ :,0:],axis=0)
print(yeary_total)
```

```python
# minimum sales per restaurant

min_sales=np.min(sales_data[:,1:], axis=1)
print(min_sales)
```

```python
#  maximum sales per year

max_sales=np.max(sales_data[:,1:], axis=0)
print(max_sales)
```

```python
# average sales per year

avg_sales=np.average(sales_data[:,1:], axis=0)
print(avg_sales)
```

```python
cumsum=np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)


# Plot average cumulative sales
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0), marker="o", linestyle="-", label="Average Sales")
plt.title("Average Cumulative Sales Across All Restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```
