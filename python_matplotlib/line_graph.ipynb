```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
month = np.array(["jan", "feb", "mar", "apr", "may", "june", "july", "aug", "sep", "oct","nov", "dec"])
rain=np.array([60, 55, 75, 80, 70, 45, 78, 55, 67, 34, 78, 46])

plt.plot(month, rain)

plt.xlabel("month")
plt.ylabel("rainfall ( mm )")

plt.title("Rainfall in an year")

plt.show()
```
