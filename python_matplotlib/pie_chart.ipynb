```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
cricketer = np.array(["Tim", "john", "Amit", "jacob", "karl", "Gary"])
runs= np.array([85, 59, 99, 77, 67, 95])

plt.pie(runs, labels=cricketer, autopct='%1.3f%%')

plt.title("Run scored in a match")
plt.show()
```

```python

cricketer = np.array(["Tim", "john", "Amit", "jacob", "karl", "Gary"])
runs= np.array([random.choice([85, 59, 99, 77, 67, 95]) for _ in range(len(cricketer))])

plt.pie(runs, labels=cricketer, autopct='%1.3f%%')

plt.title("Run scored in a match")
plt.show()
```
