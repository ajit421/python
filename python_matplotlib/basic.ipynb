```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

a = np.arange(5)
b = [2,4,6,8,10]
c = [5,6,7,8,9]

# print(a,b,c)

fig, ax = plt.subplots()

ax.plot(a, b, 'k--', label='frequency') 
ax.plot(a, c, 'k:', label='periods')

ax.legend() # display legend

ax.set_title("frequency of a signal ")
plt.show()
```

```python
fig, ax= plt.subplots()

ax.plot(a, b, 'k--', label= 'frequency')
ax.plot(a, c, 'k:', label= ' periods')

ax.legend(loc='lower right')
# 'upper left', 'upper right', 'lower left', 'lower right'
# 'upper center', 'lower center', 'center left', 'center right'


plt.title("frequency of a signal")

plt.show()
```

```python
fig,ax=plt.subplots()

ax.plot(a, b, 'k--', label='frequency')
ax.plot(a, c, 'k:', label='periods')

l=ax.legend(loc='upper center')
l.get_frame().set_facecolor('yellow')

plt.title("Frequency of a signal")

plt.show()
```

```python

fig,ax=plt.subplots()

ax.plot(a, b, 'k--', label="Frequency")
ax.plot(a,c,'k:', label="periods")

l=ax.legend(loc='upper center', fontsize='large')
l.get_frame().set_facecolor('blue')

plt.title("Frequency of a signal")

plt.show()
```
