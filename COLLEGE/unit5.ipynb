```python
f = open("sample.txt", "r")
print(f.tell())    # Position before reading
print(f.read(5))   # Reads 5 characters
print(f.tell())    # Position after reading
f.seek(0)          # Move pointer back to start
f.close()
```

```python
# Convert integer to bytes before writing
num = 1000
with open("num.bin", "wb") as f:
    f.write(num.to_bytes(2, 'big'))  # 2 bytes, big-endian format
```

```python
with open("data.bin", "rb") as f:
    content = f.read()
    print(content)
```

```python
with open("num.bin", "rb") as f:
    data = f.read()
    num = int.from_bytes(data, 'big')
    print(num)
```

```python
# Copy image using binary read and write
with open("source.png", "rb") as src:
    with open("copy.png", "wb") as dest:
        dest.write(src.read())

print("Image copied successfully!")
```

```python
import pickle

# Python dictionary
student = {"name": "Ajit", "age": 21, "course": "Python"}

# Write object to binary file
with open("student.pkl", "wb") as f:
    pickle.dump(student, f)

print("Object saved (pickled) successfully!")
```

```python
import pickle

# Read object from binary file
with open("student.pkl", "rb") as f:
    data = pickle.load(f)

print("Unpickled Data:", data)
```

```python
import pickle

data1 = [1, 2, 3]
data2 = {"x": 10, "y": 20}

with open("multi.pkl", "wb") as f:
    pickle.dump(data1, f)
    pickle.dump(data2, f)

# Reading them back
with open("multi.pkl", "rb") as f:
    a = pickle.load(f)
    b = pickle.load(f)

print(a)
print(b)
```

```python
import csv

# Open file in write mode
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age", "Course"])
    writer.writerow(["Ajit", 21, "Python"])
    writer.writerow(["Ravi", 22, "Java"])

print("CSV file written successfully!")
```

```python
data = [
    ["Name", "Age", "Course"],
    ["Neha", 20, "AI"],
    ["Kiran", 23, "ML"]
]

with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```
