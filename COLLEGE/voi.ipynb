```python
import pyttsx3

# Initialize engine
engine = pyttsx3.init()

# Voice properties
engine.setProperty('rate', 150)   # Speed
engine.setProperty('volume', 1.0) # Volume (0.0 to 1.0)

# Get available voices
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  # Male/Female voice (index may vary)

# Text to speech
engine.say("Namaste Ajit, kaise ho?")
engine.runAndWait()
```
