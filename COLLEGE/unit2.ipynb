# print multiplacation table

```python
a = int(input("Enter a value do u want to a multiplacation table:"))

for i in range(1, 11):
    print(f"{a} * {i} = {a*i}")
```

```python
def function_name():
    a = 5
    table_lines = [] # Create an empty list to store the lines
    for i in range(1, 11):
        table_lines.append(f"{a} * {i} = {a * i}") # Add each formatted line to the list
    return table_lines # Return the list of strings

for line in function_name(): # Iterate and print each line from the returned list
    print(line)
```

# Built-In Functions

```python
# No import needed for these functions
message = "Hello, Python!"

# Using print() to display output
print(message)

# Using len() to get the length of the string
length_of_message = len(message)
print("The length of the message is:", length_of_message)

# Using type() to check the data type
print("The type of 'message' is:", type(message))
print("The type of 'length_of_message' is:", type(length_of_message))

# Using input() and int()
user_age_str = input("Please enter your age: ")
user_age_int = int(user_age_str) # Convert the input string to an integer
print("Next year, you will be:", user_age_int + 1)
```

```python
import statistics

data = [10, 20, 30, 40, 50]
print(statistics.mean(data))   # 30
print(statistics.median(data)) # 30
print(statistics.stdev(data))  # Standard deviation
```

```python
import re

text = "My number is 6205607900"
pattern = r"\d+"
result = re.findall(pattern, text)
print(result)   # ['6205607900']
```

```python
def add(a, b):
    return a + b


result = add(10, 5)
print(result)   # 15
```

```python
def info(name, age):
    print(f"Name: {name}, Age: {age}")

info(age=21, name="Ajit")
```

```python
def details(**info):
    print(info)

details(name="Ajit", age=21)
```

```python
def add_all(*numbers):
    return sum(numbers)

print(add_all(2, 4, 6, 8))  # 20
```
