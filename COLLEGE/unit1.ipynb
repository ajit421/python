```python
x = 100
y = 100
z = 257
q = 257

print(f"ID of x: {id(x)}")
print(f"ID of y: {id(y)}") # Likely the same as x
print(f"ID of z: {id(z)}")
print(f"ID of q: {id(q)}") # Likely different from z (depends on Python version/implementation)

# Comparing identities
print(f"Do x and y point to the same object? {x is y}") # True
```

```python
name = "Alice"
age = 30
pi_value = 3.14
my_list = [1, 2, 3]
is_active = True

print(f"The type of name is: {type(name)}")
print(f"The type of age is: {type(age)}")
print(f"The type of pi_value is: {type(pi_value)}")
print(f"The type of my_list is: {type(my_list)}")
print(f"The type of is_active is: {type(is_active)}")
```
