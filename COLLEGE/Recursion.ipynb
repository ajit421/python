```python
def print_pattern(n):
    if n > 5:
        return
    
    print(n)
    print_pattern(n + 1)

    if n < 5:
        print(n)

print(print_pattern(1))
```

```python
def print_pattern(n, max_n=5):
    if n > max_n:
        return
    
    print(n)            # Print current number before recursion
    print_pattern(n + 1, max_n)  # Recursive call with next number
    print(n)            # Print current number again on return (unwinding)

print_pattern(1)
```

## factorial

```python
def factorial(n):
    if n ==0 or n ==1:
        return 1
    return n * factorial(n-1)
print(factorial(5)) 
```

### fibonacci

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)
print(fib(4))
```

```python
def fab(num):
    if num==0:
        return 0
    elif num==1:
         return 1

    else:
         return fab(num-1)+fab(num-2)
print(fab(5))
```

### palindrome


```python
def palindrome(n):
    rev = str(n)[::-1]
    return rev == str(n)

print(palindrome("ajija"))
```
